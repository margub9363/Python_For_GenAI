```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
paragraph = """
“I have three visions for India. In 3000 years of our history, people from all over the world have come and invaded us…
The Greeks, the Turks, the Moguls, the Portuguese, the British…
They all conquered us, took over what we had built. But we have never invaded any other nation.”
"""
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
# Apply stopwords and filter and then apply Stemming

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    [stemmer.stem(word) for word in words if word not in  set(stopwords.words('english'))] 
    sentences[i] = ' '.join(words)
```
